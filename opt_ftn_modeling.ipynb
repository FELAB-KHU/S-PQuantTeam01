```python
# 'priceClose' 컬럼에서 NaN 값을 찾습니다.
nan_rows = new_mkt[new_mkt['priceClose'].isna()]

# 결과를 출력합니다.
nan_rows
```

```python
nan_rows.info()
```

```python
import yfinance as yf


# nan 값을 수정 종가로 채우는 함수
def fill_nan_with_yfinance(nan_rows):
    # 수정된 종가를 저장할 리스트를 초기화합니다.
    adjusted_closes = []
    
    # nan_rows 데이터프레임의 각 행에 대해 반복합니다.
    for index, row in nan_rows.iterrows():
        ticker = row['tickerSymbol']
        date = row['pricingDate']
        # yfinance를 통해 해당 티커의 데이터를 가져옵니다.
        data = yf.download(ticker, start=date, end=pd.to_datetime(date) + pd.Timedelta(days=1))
        
        # 데이터가 비어 있지 않다면, 수정 종가를 가져와 리스트에 추가합니다.
        if not data.empty and 'Adj Close' in data.columns:
            adjusted_close = data['Adj Close'].iloc[0]
            adjusted_closes.append(adjusted_close)
        else:
            # 데이터가 없을 경우 None을 추가합니다.
            adjusted_closes.append(None)
    
    # nan_rows의 'priceClose' 컬럼에 수정 종가를 채워줍니다.
    nan_rows['priceClose'] = adjusted_closes
    return nan_rows

# 함수를 호출하여 NaN 값을 수정 종가로 채워줍니다.
filled_nan_rows = fill_nan_with_yfinance(nan_rows)
filled_nan_rows
```
