```python
mkt
```

```python
mkt_symbol = list(mkt['tickerSymbol'].unique())
mkt_symbol  # market data에 있는 티커
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
sp500_table = pd.read_html(url, header=0)[0]
tickers = sp500_table['Symbol'].tolist()
tickers, len(tickers)  # S&P500 구성종목 티커
```

```python
# 두 리스트를 집합으로 변환하고 교집합을 구함
intersection = list(set(tickers) & set(mkt_symbol))

# 결과 출력
intersection, len(intersection)
```

```python
# mkt 데이터프레임에서 tickerSymbol이 intersection에 있는 행만 필터링
new_mkt = mkt[mkt['tickerSymbol'].isin(intersection)]
new_mkt
```

```python
new_mkt['adjClose'] = new_mkt['priceClose'] * new_mkt['adjustmentFactor'] * new_mkt['divAdjFactor']
new_mkt
```

```python
# divAdjFactor가 None에 해당하는 행 필터링 및 tickerSymbol 추출
none_divAdjFactor_tickers = new_mkt[new_mkt['divAdjFactor'].isnull()]['tickerSymbol']

# 결과 출력
set(none_divAdjFactor_tickers), len(set(none_divAdjFactor_tickers))
```
