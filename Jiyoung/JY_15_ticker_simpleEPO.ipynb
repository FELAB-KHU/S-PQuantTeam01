```python
import pandas as pd
```

```python
ticker_15 = pd.read_csv('./all_prices_15.csv', index_col= 'Date')
ticker_15.index = pd.to_datetime(ticker_15.index)
ticker_15
```

```python
## 공분산행렬
cov_matrix = ticker_15.cov()
cov_matrix
```

```python
# 가중치 제약
n_assets = cov_matrix.shape[0]
min_weight = (1/2) * (1/n_assets)
max_weight = 2 * (1/n_assets)
```

```python
# 15 종목에 대해 opt
import numpy as np
from scipy.optimize import minimize

# 목적함수
def portfolio_variance(weights, cov_matrix):
    return np.dot(weights.T, np.dot(cov_matrix, weights))

# 상관관계 축소를 위한 λ 값 설정 (예시로 0.75를 사용)
lambda_shrinkage = 0.75

# 대칭 행렬로 조정
cov_matrix_np = cov_matrix.to_numpy()
cov_matrix_symmetric = (cov_matrix_np + cov_matrix_np.T) / 2

# 상관관계 행렬 조정
corr_matrix = ticker_15.corr()  # 상관관계 행렬 계산
identity_matrix = np.eye(corr_matrix.shape[0])
adjusted_corr_matrix = (1 - lambda_shrinkage) * corr_matrix + lambda_shrinkage * identity_matrix

# 각 자산의 변동성 계산
volatilities = np.sqrt(np.diag(cov_matrix_symmetric))

# 분산-공분산 행렬 재계산
cov_matrix_psd = np.outer(volatilities, volatilities) * adjusted_corr_matrix

# 제약조건
constraints = ({'type': 'eq', 'fun': lambda x: np.sum(x) - 1})
bounds = [(min_weight, max_weight) for _ in range(n_assets)]

# 초기 추정치
initial_guess = [1/n_assets] * n_assets

# 최적화
result = minimize(portfolio_variance, initial_guess, args=(cov_matrix_psd,), method='SLSQP', bounds=bounds, constraints=constraints)

# 종목수 컬럼
ticker_symbols = cov_matrix.columns

# 최적 가중치
optimal_weights = result.x
optimal_weight_dict_15 = {ticker: weight for ticker, weight in zip(ticker_symbols, optimal_weights)}
```

```python
# 딕셔너리를 데이터 프레임으로 변환
ticker_15_optimal_weights = pd.DataFrame(list(optimal_weight_dict_15.items()), columns=['Ticker', 'Weight'])
ticker_15_optimal_weights.to_csv('./ticker_15_EPO_optimal_weights.csv', index=False)
```
