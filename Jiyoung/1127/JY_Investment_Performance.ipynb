```python
#SnP500 주간수익률 계산

from pandas_datareader import data as pdr 
import yfinance as yf 
yf.pdr_override()

def yahoo_reader(ticker, start='2018-04-24', end='2023-04-30', freq=None, calc_return=True, calc_log=False):
    """Read daily data from Yahoo finance
    Assume frea is valid (e.g., 'M', 'W-FRI') 
    calc_return: calculate return (adjusted return) 
    calc_log: calculate log of data
    """
    df = pdr.get_data_yahoo(ticker, start=start, end=end).fillna(method='ffill')['Adj Close']
    
    if freq is not None:
        # Apply resample frequency is necessary
        df = df.resample(freq).last()
    
    if calc_log:
        df = np.log(df)
    if calc_return:
        df = df.pct_change().fillna(0.0)
    
    df = df. dropna()
    df = df[df.index >= start] # in case previous date is included
    
    return df
```

```python
# 성과평가 함수

from scipy.stats import norm
def ret_mean(returns, dt=1): #weekly를 annual로 바꾸고싶으면 dt=52
    return np. mean(returns, axis=0) * dt

def ret_std(returns, dt=1):
    return np.std(returns, ddof=1, axis=0) * np.sqrt(dt)

def total_return(returns):
    return (1 + pd. DataFrame(returns)). prod() - 1

def mdd(returns):
    df_wealth = (pd. DataFrame(returns) + 1). cumprod(axis=0)
    mdd = (df_wealth / df_wealth. expanding(min_periods=0). max()) .min() - 1
    return mdd

def value_at_risk(returns, conf=0.95, assume_norm=False):
    """Value-at-risk either under normality assumption or not 
    Must enter a single column for 'returns'
    """
    if assume_norm:
        # Assume normal distribution of returns
        var = norm. ppf(1 - conf, np. mean(returns), np.std(returns, ddof=1))
    else:
        # No assumption on return distribution
        returns_sorted = returns. sort_values(). reset_index(drop=True)
        i = int(np.floor((1 - conf) * len(returns_sorted)))
        var = returns_sorted[i - 1]
    return var

def conditional_value_at_risk(returns, conf=0.95, assume_norm=False):
    """Conditional value-at-risk either under normality assumption or not
    Must enter a single column for 'returns'
    """
    var = value_at_risk(returns, conf=conf, assume_norm=assume_norm)
    return np. mean(returns[returns <= var])

def cvar(returns, conf=0.95, assume_norm=False):
    """
    Shorthand for conditional_value_at_risk()
    """
    return conditional_value_at_risk(returns, conf=conf, assume_norm=assume_norm)
```
